```python
import pandas as pd
```

```python
# Memuat Data
df = pd.read_csv('Data Mining/movie_sample_dataset.csv')
df.head(10)
```

```python
# Memeriksa Data
df.dtypes
```

```python
# Memeriksa jumlah missing values di setiap kolom
print(df.isnull().sum())
```

```python
import numpy as np
```

```python
# Membersihkan Data
# Mengganti "N/A" dengan NaN
df.replace("N/A", np.nan, inplace=True)
```

```python
# 1. Menghapus baris yang memiliki NaN di kolom 'gross' dan 'budget'
# Menampilkan jumlah nilai NaN sebelum penghapusan
print("Jumlah nilai NaN sebelum penghapusan:")
print(df.isnull().sum())
```

```python
# Menghapus baris yang memiliki NaN di kolom 'gross' dan 'budget'
df_cleaned = df.dropna(subset=['gross', 'budget']).copy()
```

```python
# Menampilkan jumlah nilai NaN setelah penghapusan
print("\nJumlah nilai NaN setelah penghapusan:")
print(df_cleaned.isnull().sum())
```

```python
# Menampilkan beberapa baris dari DataFrame yang sudah dibersihkan
print("\nMenampilkan 5 baris pertama dari DataFrame yang sudah dibersihkan:")
print(df_cleaned.head(10))
```

```python
# 2. Menangani kesalahan penulisan di kolom 'color'
# Mengganti variasi 'color' menjadi 'Color'
df_cleaned['color'] = df_cleaned['color'].replace({'color': 'Color', 'colour': 'Color'})

# Menangani kesalahan penulisan di kolom 'country' (contoh)
# Mengubah 'usa' menjadi 'USA' dan 'uk' menjadi 'UK'
df_cleaned['country'] = df_cleaned['country'].replace({'usa': 'USA', 'uk': 'UK'})

# Normalisasi teks untuk kolom yang diinginkan
df_cleaned['color'] = df_cleaned['color'].str.capitalize()  # Mengubah ke format Capitalized
df_cleaned['country'] = df_cleaned['country'].str.upper()  # Mengubah ke huruf kapital semua

# Menampilkan hasil akhir
print(df_cleaned.head(10))
```

```python
# 3. Ubah atau hapus nilai-nilai yang tidak standar, seperti nilai negatif atau "N/A".
# Ubah semua nilai negatif di seluruh DataFrame menjadi NaN
# Ubah nilai negatif di kolom 'duration' menjadi NaN
df['duration'] = np.where(df['duration'] < 0, np.nan, df['duration'])

# Jika kamu ingin menghapus semua baris yang mengandung NaN setelah penggantian
df.dropna(inplace=True)

# Menampilkan hasil akhir
print("DataFrame setelah menghapus nilai negatif dan 'N/A':")
print(df.head(10))
```

```python
# Transformasi Data
# 1. Mengubah tipe data kolom gross dan budget
df[["gross", "budget"]] = df[["gross",
"budget"]].astype(int)

# Menghapus baris dengan NaN yang mungkin muncul setelah konversi tipe data
df.dropna(subset=['budget', 'gross'], inplace=True)

# Menampilkan hasil akhir
print("DataFrame setelah mengubah tipe data 'budget' dan 'gross':")
print(df.head())
```

```python
# 1. Mengubah tipe data kolom 'budget' dan 'gross' menjadi numerik
# Gunakan pd.to_numeric() untuk mengonversi dan atasi error jika ada nilai yang tidak bisa dikonversi
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df['gross'] = pd.to_numeric(df['gross'], errors='coerce')

# Menghapus baris dengan NaN yang mungkin muncul setelah konversi tipe data
df.dropna(subset=['budget', 'gross'], inplace=True)

# 2. Memisahkan kolom 'genre' yang memiliki beberapa nilai dalam satu baris
# Misalkan nilai-nilai dalam kolom 'genre' dipisahkan dengan koma
df_genres = df['genres'].str.split(',', expand=True)

# Gabungkan kembali ke DataFrame asli (opsional)
# Ini akan menambahkan kolom-kolom baru sesuai dengan jumlah genre yang ditemukan
df = pd.concat([df, df_genres], axis=1)

print("Menampilkan 5 baris pertama dari DataFrame:")
print(df.head())  # Menampilkan 5 baris pertama
```

```python
# Menyimpan Data
df.to_csv('movie_dataset_cleaned.csv', index=False)
```
